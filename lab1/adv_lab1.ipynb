```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# dataset :https://www.kaggle.com/datasets/anuchhetry/product-sales
```

```python
df = pd.read_excel('data.xlsx')
```

```python
df.columns
```

```python
df.head()
```

### Bar Plot

```python
# Bar Plot

sns.set(style="whitegrid")
palette = sns.color_palette("husl")
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Product', y='Sale Price', hue='Segment')
plt.title('Sale Price by Product and Segment', fontsize=16)
plt.xlabel('Product', fontsize=14)
plt.ylabel('Sale Price', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Segment', title_fontsize='13', fontsize='11', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

### Pie Chart

```python
# Pie Chart

colors = sns.color_palette("coolwarm", len(df['Product'].unique()))
plt.figure(figsize=(10, 8))
df['Product'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',  # Show percentages with one decimal place
    startangle=140,  # Start the first slice at a specified angle
    colors=colors,
    wedgeprops={'edgecolor': 'black', 'linewidth': 1.5},  # Add a border to the slices
    textprops={'fontsize': 14, 'color': 'black'},  # Increase font size of labels
    shadow=True,  # Add a shadow effect for depth
    explode=[0.05] * len(df['Product'].unique())  # Slightly "explode" all slices for emphasis
)
plt.title('Product Distribution', fontsize=20, fontweight='bold')
plt.gca().set_aspect('equal')
plt.ylabel('')
plt.show()
```

### Hisplot

```python
# Histplot

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(df['Profit'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Profit', fontsize=16, fontweight='bold')
plt.xlabel('Profit', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.show()
```

### Timeline Plot

```python
# Timeline Plot

sns.set_theme(style="darkgrid", rc={'figure.figsize': (11.7, 8.27)})
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="Date", y="Profit")
plt.title('Timeline of Profit', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Profit', fontsize=14)
plt.show()
```

### Scatter Plot

```python
# Scatter Plot

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Gross Sales", y="Profit")
plt.title('Scatter Plot of Gross Sales vs Profit', fontsize=16)
plt.xlabel('Gross Sales', fontsize=14)
plt.ylabel('Profit', fontsize=14)
plt.show()
```

### Bubble Plot

```python
# Bubble Plot

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Units Sold", y="Profit", size="Sale Price", hue="Segment", alpha=0.7)
plt.title("Bubble Plot of Units Sold vs Profit, Sized by Sale Price", fontsize=16)
plt.xlabel("Units Sold", fontsize=14)
plt.ylabel("Profit", fontsize=14)
plt.show()
```
